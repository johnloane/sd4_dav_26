**Clustering**
- Machine learning technique that uses unsupervised labels
- Group unlabelled data into natural clusters
- Clustering is optimisation problem
- $variability(c) = \sum_{e \in c}distance(mean(c), e)^2$
- Variance is different from variability, you don't divide by n. This is because in clustering big and bad is worse than small and bad. We don't want to average out the difference
- $dissimilarity(C) = \sum_{c \in C}variability(c)$
- Find C that minimises the dissimilarity
- In order for this not end up as each point being a cluster, we add a constraint e.g the clusters have to have a minimum distance or constrain the number of clusters

**Hierarchical clustering**
- Start by assigning each item to a cluster so that if you have N items, you now have N clusters, each containing one item
- Find the closest (most similar) pair of clusters and merge them into a single cluster, so now we have one fewer clusters
- Continue until all items are clustered in a single cluster of size N
- Agglomerative hierarchical clustering
- What do we mean by closest?
- Can get clusters depending on the linkage metrics
- **Single linkage**: consider the distance between one cluster and another to be equal to shortest distance between any member of one cluster and any member of the other cluster
- **Complete linkage**: consider the distance between one cluster and another as the greatest from any member of one cluster to any member of the other cluster
- **Average linkage**: consider the distance between one cluster and another as the average distance from any member of one cluster to any member of the other cluster

**Pros and cons of hierarchical**
- Get history of dendrograms
- Deterministic given a linkage metric
- Answer might not be optimal. Greedy algorithm. Making locally optimal decisions at each point that might not be globally optimal
- Flexible
- Really slow. Complexity $O(n^3)$. Not something to use with large datasets

**k-means clustering**
- k is the number of clusters that you want
- If you don't have any idea about how many clusters there are this can be problematic
- If you do know how many clusters there are this is a good choice as it is very fast


**k-means algorithm**
- Randomly choice sample as initial centroid
- while True:
- create k clusters by assigning each point to the closest centroid
- compute k new centroids by averaging the examples in each cluster
- if centroids don't change we are finished

**k means complexity**
- Complexity of one iteration
- Typically converges quite quickly
- Most commonly used clustering algorithm because of its speed


**Choosing k problems**
- Choosing wrong k can be problematic
- If k is non-sensical then your clustering will be con-sensical
- How do we choose k?
- Results can depend on the initial centroids
- k-means is non-deterministic
- If the initial centroids are poorly chosen it can take longer to converge or you can get a completely different answer

**Choosing k solutions**
- Choose k with a-priori knowledge of the application
- For example there are only 5 different types of bacteria
- Two types of people in the world - DkIT graduates and not
- Try different values of k and evaluate the quality of the results
- Run hierarchical clustering on a subset of the data
- Try and select good initial centroids distributed over the space
- Try multiple sets of random chosen centroids and choose the best result


**Mitigating dependence on initial centroids**
- Best = kmeans(points)
- for t in range(numTrials):
- C = kmeans(points)
- if dissimilarity(C) < dissimilarity(best)
- best = C
- return C
- Think of the centroid as the "centre of gravity" of a cluster


**Elbow method to find k**
- $SSE_1 = \sum_{i=0}^n dist(x_i - c_1)^2$
- $SSE = SSE_1 + \dots + SSE_n$ 

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("income.csv")
```

```python
df.head()
```

```python
plt.scatter(df.Age, df.Income)
```

```python
km = KMeans(n_clusters=3, n_init='auto')
```

```python
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age, df0.Income, color='green')
plt.scatter(df1.Age, df1.Income, color='red')
plt.scatter(df2.Age, df2.Income, color='black')
```

- Problematic as features are not scaled
- Income 40,000 -> 160,000
- Age 25 - 45
- Income will completely dominate unless we scale. If you don't scale the data you might as well forget about age because income is 2000 times what age is

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income']])
df.Income = scaler.transform(df['Income'].values.reshape(-1,1))
```

```python
df
```

```python
scaler.fit(df[['Age']])
df.Age = scaler.transform(df['Age'].values.reshape(-1,1))
```

```python
df
```

```python
km = KMeans(n_clusters=3, n_init='auto')
y_predicted = km.fit_predict(df[['Age', 'Income']])
df.cluster = y_predicted
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age, df0.Income, color='green')
plt.scatter(df1.Age, df1.Income, color='red')
plt.scatter(df2.Age, df2.Income, color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', marker='*', label='centroid')
```

```python
km.cluster_centers_
```

```python
k_range = range(1, 10)
sse = []
for k in k_range:
    km = KMeans(n_clusters = k, n_init='auto')
    km.fit(df[['Age', 'Income']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel("Sum of squared error")
plt.plot(k_range, sse)
```
