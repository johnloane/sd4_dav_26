**New Considerations**
1. Add more independent variables to multiple linear progression procedure does not mean that it will be "better" or offer better predictions. In fact, it can make things worse. This is called **overfitting**
2. The addition of more independent variables creates more relationships among variable. So not only are the independent variables related to the dependent variable, they are also related to each other. When this happens it is called multi-colinearity.
3. The ideal is for all of the independent variables to be correlated with the dependent variable but **not** with each other.
4. Because of multi-colinearity and overfitting, there is a fair amount of prep-work to do before conducting multiple linear regression.
- Correlations
- Scatterplots
- Simple regressions between each independent variable and the dependent variable

-If two independent variables are related to each other we are not sure which one is explaining the variation in the dependent variable.
-If we put sea salt and table salt into our dinner and our dinner is salty we don't know which one is driving the salty taste.
- $n(n-1)/2$ will tell us the number of relationships between the independent variables. There are are addition $n$ relationships between the independent variables and the dependent variable
- Some independent variables, or sets of independent variables, are better at predicting the dependent variable than others. Some contribute nothing.


***Multiple linear regression model

$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_px_p + \epsilon$

Multiple regression equation:

$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_px_p$
error term is assumed to be zero

Estimated multiple regression equations:

$ \hat{y} = b_0 + b_1x_1 + b_2x_2 + \dots + b_px_p$

$\hat{y}$ is the predicted value of the dependent variable. $b_0 \dots b_p$ estimates of $\beta_0 \dots \beta_p$

$ \hat{y} = 6.211 + 0.014x_1 + 0.383x_2 - 0.607x_3$

$\hat{y} = 27 + 9x_1 + 12x_2$
- $x_1$ =  capital investment (€1000's)
- $x_2$ = marketing expenditure (€1000's)
- $\hat{y}$ = predicted sales (€1000's)
- Each coefficient is interpreted as the estimated change in $y$ corresponding to a 1 unit change in a variable when all other variables are held fixed.

**Review**
1. Multiple regression is an extension of linear regression
2. Two or more independent variables are used to predict/explain the variance in one dependent variable
3. Two problems that arise- overfitting and multi-colinearity
4. Overfitting is caused by adding too many independent variables. They account for more of the variance but they add nothing to the model
5. Multi-colinearity happens when some/all of the independent variables are correlated to each other
6. In multiple regression each coefficient is interpreted as the estimated change in $y$ corresponding to a one unit change in a variable, when all the other variables are kept fixed

$R^2$ is your go to metric for seeing how well the model fits the data.
1. $R^2$ measures the proportion of the variance in the dependent variable that is predictable from the independent variables.
- What is tells you: "How much of the variance in my data did I explain?"
- Flaw: $R^2$ will never decrease when you add more variables, even if the variables are complete nonsense. This can lead to overfitting.
2. Adjusted $R^2$ compensates for the number of variables in your model. It only increases if the new variable improves the model more than would be expected by chance.
- What is tells you: "Is the extra variable useful"
- Key difference: Adjusted $R^2$ can decrease if you add useless independent variables. It is a more honest metric when comparing models with different numbers of parameters.
3. Predicted $R^2$. While the first two look at well the model fits the data you already have, predicted $R^2$ tells you how well the model works with new observations. 
- How it works: It is calculated by systematically removing each point from the set, refitting the model, and seeing how well the model predicts the "missing" point (cross validation)
- What it tells you: "Can I trust the model in the real world, or did it just memorise the current dataset?"
- Red Flag: If your $R^2$ is 0.9 and $R^2$ predicted is 0.2 you model it severely overfit. It is great at explaining the past but useless for explaining the future

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
km_travelled = [89, 66, 78, 111, 44, 77, 80, 66, 109, 76]
num_deliveries = [4, 1, 3, 6, 1, 3, 3, 2, 5, 3]
petrol_prices = [3.84, 3.19, 3.78, 3.89, 3.57, 3.57, 3.03, 3.51, 3.54, 3.25]
df = pd.DataFrame({
    'km_travelled' : km_travelled,
    'num_deliveries' : num_deliveries,
    'petrol_prices' : petrol_prices
})
X = df[['km_travelled', 'num_deliveries']]
X = add_constant(X)
#Calculate VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range (len(X.columns))]
print(vif_data)

```

VIF detects multi-colinearity. It measures how much the variance in an estimated regression coefficient is increased due to multi-colinearity.
The VIF tells you how much of the signal in one variable is actually just a reflection of the others.
- VIF = 1: No correlation between independent variables
- VIF between 1 and 5: Moderate correlation, usually acceptable
- VIF > 5, > 10: High correlation. You coefficients become unreliable and p-values might be misleading.

```python
# Mallows Cp
import pandas as pd
import statsmodels.api as sm

df = pd.DataFrame({
    'km_travelled' : km_travelled,
    'num_deliveries' : num_deliveries,
    'petrol_prices' : petrol_prices,
    'time_taken' : time_taken
})

# Define the full model (all independent variables)
# Use MSE from this model as our true estimate
X_full = sm.add_constant(df[['km_travelled', 'num_deliveries', 'petrol_prices']])
y = df['time_taken']
full_res = sm.OLS(y, X_full).fit()
mse_full = full_res.mse_resid
#Define the subset model
X_sub = sm.add_constant(df['km_travelled'])
sub_res = sm.OLS(y, X_sub).fit()
#Calculate Mallows Cp
rss_sub = sub_res.ssr
n = len(y)
p = len(sub_res.params)
mallows_cp = (rss_sub/mse_full)-n+(2*p)
print(f"Mallows Cp for the subset model: {mallows_cp:.4f}")
print(f"Number of parameters (p): {p}")
```

Mallows $C_p$ is a technique for model selection to identify which subset of independent variables creates the best model without overfitting
- It compares the full model against a smaller reduced model
- The goal is to find a model that is small enough to be easy to interpret but large enough to capture the actual relationships between variables. 
- With Mallows $C_p$ we are looking for a relationship between the $C_p$ vlaues and the number of independent variables:
- Ideal Model: $C_p \approx p+1$
- Overfit model: $C_p < p+1$
- Underfit model: $C_p > p+1$


```python
km_travelled = [89, 66, 78, 111, 44, 77, 80, 66, 109, 76]
num_deliveries = [4, 1, 3, 6, 1, 3, 3, 2, 5, 3]
petrol_prices = [3.84, 3.19, 3.78, 3.89, 3.57, 3.57, 3.03, 3.51, 3.54, 3.25]
data = list(zip(km_travelled, num_deliveries, petrol_prices))
print(data)
data = sm.add_constant(data)
model = sm.OLS(time_taken, data)
result = model.fit()
print(result.summary())
# Print the standard error
print(result.scale**0.5)
# Work out R-squared predicted
influence = result.get_influence()
#Calculate PRESS statistic
press = np.sum((result.resid/(1-influence.hat_matrix_diag))**2)
#calculate predicted R squared
sst = result.centered_tss
predicted_r2 = 1-(press/sst)
print(f"Predicted r squared {predicted_r2:.4f} ")
```

**Total sum squares**
-SST is the sum of the squared distances from the mean model ( a horizontal line having slope = 1 at the average point) and each observation.
- Once the data is located in w.r.t the total number of observations (rows) the total sum of squares is fixed
- SST is a finite sum that does not change, regardless of how many features we put into the model, SST is not feature dependent, it is row dependent
- Independent variables partition the SST between the model and an error term
- As features are added and removed the partitioning of SSR and SSE will almost certainly
- How the sum of squares is allocated to each feature with SSR will change as the model changes
- If no independent variables have significant explanatory power on price, the model will have zero features, and the best prediction is the just the average


SST is split into:
- SSR (regression model sum of squares)
- SSE (residual error sum of squares)
$R^2 = SSR/SST$
$R^2$ is the proportion of the sum of squares that we can put in the model
The goal is to
- Maximise SSR as proportion of SST
- Minimise SSE as proportion of SST
- SSR is then split among the features in the model in terms of relative explanatory power
- Not only is the SSR likely to change but the allocation of the SSR to each feature in the model will change as features are added and removed

**Overfitting and SSR**
- As more features are added to the model SSR will never decrease; SSR can only stay the same or increase
- As additonal features are added, the model to shape itself to the data it is trained on, keeps on optimising
- The model seeks to squeeze out as much SSE as possible whether the feature adds predictive power or not
- Complex regressions could suffer on new data because of the nooks and crannies of the training data are absent
- There are techniques to avoid overfitting:
1. Model bulding procedures (Forward selection, backward elimination, stepwise regression and best subsets regression) along with AIC, BIC, and Mallows $C_p$
2. Regularization such L1 and L2 penalize complex models
3. Cross validation
4. Principal Component Analysis

**Conclusion**
- Total sum of squares (SST) is fixed once the rows/data is fixed
- Sum of the squares for the model(SSR) will never go down as more variables are added
- Sum of squares error (SSE) will never go up as more variables are added
- Adding more features will either have no effect or reduce SSE
- Regression will add all features unless stopped by a rule
- Dumping in all features can lead to overfitting an unnecessary complexity
- Features can be added more than one at a time so long as all the features in the simple model and in the complex model


**Multiple Regression - Forward selection**
There are three common techniques for building iterative regression models:
1. Forward selection
2. Backward elimination
3. Stepwise regression
The fourth common technique is not iterative. It is called best subsets regression.
- Best subsets examines all possible combinations of feature variables; brute force method that can computationally expensive.
- For example small model with 9 feature variables
- 9, 36, 84, .......
- Best subsets will examine all models - hundreds
- You can specify the maximum number of features
- You can also request the best two or three models for each number of variables

**Forward selection**
- Feature variables are examined on at a time
- The analyst sets a stopping rule (p-value) that acts as a hurdle the variable must overcome to get into the model
- The added variable must reduce the error significantly
- The feature variable that reduces the model error (SSE) the most is chosen so long as it passes the stopping rule
- Once a variable is in the model, it stays. It is never removed.
- Then the process repeats until all variables are in the model or no more variables pass the stopping rule
